```python
class Employer:
  def __init__(self):
    self.staff = {
        "Manager" :{"name" : "Sindhu", "hrs_worked": 0, "salary" : 5000},
        "Cheif of cleaning": {"name" : "Andy", "hrs_worked": 0, "salary" : 4000},
        "Cheif of inventory": {"name" : "Sally", "hrs_worked": 0, "salary" : 3000},
        "Barista1": {"name" : "Quina", "hrs_worked": 0, "salary" : 750},
        "Barista2": {"name" : "William", "hrs_worked": 0, "salary" : 750},
        "Barista3": {"name" : "Grag", "hrs_worked": 0, "salary" : 750}
    }

  def hrs_worked(self,role, sign_in, sign_out):
    if role in self.staff:
      #hrs = sign_out -sign_in
      sign_in_hours = sign_in // 100
      sign_in_minutes = sign_in % 100
      sign_out_hours = sign_out // 100
      sign_out_minutes = sign_out % 100

      # Convert everything to total minutes
      total_sign_in_minutes = sign_in_hours * 60 + sign_in_minutes
      total_sign_out_minutes = sign_out_hours * 60 + sign_out_minutes

      # Calculate total minutes worked
      total_minutes_worked = total_sign_out_minutes - total_sign_in_minutes

      # Convert minutes back to hours
      worked_hours = total_minutes_worked // 60
      worked_minutes = total_minutes_worked % 60
      self.staff[role]["hrs_worked"] += worked_hours + (worked_minutes / 60)
      print(f"{self.staff[role]['name']} ({role}) worked for {worked_hours} hours")

    else:
      print("role not found")

  def display_staff(self):
    print("\n----Staff----")
    for role, deets in self.staff.items():
      print(f"{role} : {deets['name']}, Hours Worked: {deets['hrs_worked']}, Salary Rate: rs {deets['salary']}/hour ")
  def display_role_hrs(self, role):
    if role in self.staff:
      deets = self.staff[role]
      print(f"{deets['name']} ({role}): {deets['hrs_worked']} hours worked")
    else:
      print("role not found")
  def calc_payroll(self):
    tot_pay = 0
    for role, deets in self.staff.items():
      pay = deets["hrs_worked"] * deets["salary"]
      print(f"{deets['name']} ({role}): rs {pay}")
      tot_pay += pay
    print(f"\nTotal Payroll: rs {tot_pay}")
  def reset_hrs_worked(self):
    for details in self.staff.values():
        details["hrs_worked"] = 0
    print("\nHours worked have been reset for all staff.")

```

```python

class Customer:
  def __init__(self,name):
    self.name = name
    self.points = 0
    self.orders = []

  def add_points(self,points):
    self.points+=points
  def remove_points(self,points):
    if self.points >= points:
      self.points-=points
      print(f"{points} points deducted from {self.name}. Remaining points: {self.points}")
    else:
      print(f"Not enough points to deduct {points} points from {self.name}.")
  def display_points(self):
    print(f"{self.name}'s Points: {self.points}")

  def add_order(self, order):
    self.orders.append(order)

  def display_orders(self):
    print(f"{self.name}'s Order History:")
    for order in self.orders:
        print(order)
class CoffeeShop:
  def __init__(self):
    self.menu={
        'esspreso' : 100,
        'latte' : 200,
        'cappuccino' : 300,
        'macchiato' : 400,
        'americano' : 500
    }
    self.inventory={
        'esspreso' : 1000,
        'latte' : 800,
        'cappuccino' : 750,
        'macchiato' : 500,
        'americano' : 400
    }
    self.sales = 0.0
    self.customers={}
    self.sales_breakup = {item: 0 for item in self.menu}


  def display_menu(self):
    print("Menu : ")
    for item, price in self.menu.items():
      print(f"{item}: rs {price}")
  '''
  def take_order(self,item,customer_name):
    if item in self.menu:
      if self.check_inventory(item):
        self.process_order(item,customer_name)
        print(f"Enjoy your {item}, {customer_name}!")
      else:
        print("Sorry, we are out of stock for this item.")
    else:
      print("Sorry, that item is not on the menu.")
  '''

  def take_order(self, customer_name):
    ord = {}
    tot = 0
    while True:
      item = input("Enter the item: ").strip()
      if item in self.menu:
        qty = int(input(f"Qty {item}: "))
        #ord[item] = qty
        if self.check_inventory(item,qty):
          if item in ord:
            ord[item]+= qty
          else:
            ord[item] = qty
          print(f"Added {qty} {item}(s) to your order.")
        else:
          print("Sorry, we are out of stock for this item.")
      else:
        print("Sorry, that item is not on the menu.")
      others = input("Would you like to order something else? (y/n): ").lower()
      if others != "y":
          print("Processing your order...")
          break

    # Process the collected orders
    for item, qty in ord.items():
        if qty > 1:
            self.process_bulk_orders({item: qty}, customer_name)
        else:
            self.process_order(item, customer_name)

        tot += self.menu[item] * qty

    print(f"Order complete for {customer_name}. Total Bill: rs {tot}")
    print(f"Enjoy your coffee, {customer_name}!")


  def process_bulk_orders(self, ord, customer_name):
      for item, qty in ord.items():
          self.sales += self.menu[item] * qty
          self.sales_breakup[item] += qty
      self.update_customer_points(customer_name, sum(self.menu[item] * qty for item, qty in ord.items()))
      self.record_order(customer_name, ord)

    #print(f"Order processed for {customer_name}. Total: rs {tot}")

  def check_inventory(self,item, qty):

    if item in self.inventory:
      if self.inventory[item]>0 and self.inventory[item]>=qty:
        self.inventory[item]-=qty
        return True
      else:
        return False
    return False

  def process_order(self, item, customer_name):
    self.sales+=self.menu[item]
    self.update_customer_points(customer_name,self.menu[item])
    self.record_order(customer_name,{item: 1})
    self.sales_breakup[item]+=1

  def display_inventory(self):
    print("Current Inventory")
    for item, quantity in self.inventory.items():
      print(f"{item}: {quantity}")
  def replenish_inventory(self, item, quantity):
    if item in self.inventory:
      self.inventory[item] += quantity
      print(f"Added {quantity} units of {item} to inventory.")
    else:
      print("That item does not exist in the inventory.")
  def update_customer_points(self, customer_name, amount_spent):
        if customer_name in self.customers:
            self.customers[customer_name].add_points(int(amount_spent * 10))  # Reward 10 points for each rs spent
        else:
            self.customers[customer_name] = Customer(customer_name)
            self.customers[customer_name].add_points(int(amount_spent * 10))
  def record_order(self,customer_name,item):
    if customer_name in self.customers:
      self.customers[customer_name].add_order(item)
    else:
      self.customers[customer_name] = Customer(customer_name)
      self.customers[customer_name].add_order(item)

  def display_customer_order(self, customer_name):
    if customer_name in self.customers:
      self.customers[customer_name].display_orders()
    else:
      print(f"{customer_name} is not a registered customer.")

  def apply_discount(self,customer_name):
    if customer_name in self.customers:
      if self.customers[customer_name].points >=100:
        discount_percent = 10
        discount_amount = self.sales * (discount_percent / 100)
        self.sales -= discount_amount
        self.customers[customer_name].remove_points(100)
        print(f"{customer_name} received a {discount_percent}% discount! New total:  rs {self.sales:.2f}")
      else:
        print(f"{customer_name} does not have enough points for a discount.")
    else:
      print(f"{customer_name} is not a registered customer.")

  def report_sales(self):
    print("\n--- breakup of sales---")
    for item, sold in self.sales_breakup.items():
      print(f"{item}: {sold}")
    print("\n--- Daily Sales Report ---")
    print(f"Total Sales: rs {self.sales:.2f}")




  def report_inventory(self):
    total_items = sum(self.inventory.values())
    print(f"Total Inventory Items: {total_items}")
```
